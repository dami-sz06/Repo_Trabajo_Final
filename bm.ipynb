```python
import tkinter as tk
import random
import time

# --- CONSTANTES GLOBALES (est치 bien dejarlas aqu칤) ---
TAMANO_GRID = 10
NUM_BOMBAS = 10

# --- VARIABLES GLOBALES (tambi칠n est치 bien) ---
botones_grid = {} 
es_bomba = {}   
juego_terminado = False
inicio = 0 # Solo la definimos, la iniciaremos luego

# --- FUNCIONES DEL JUEGO (perfectas como est치n) ---
def casilla_presionada(r, c):
  global juego_terminado
  # ... (tu c칩digo de casilla_presionada sin cambios) ...
  if juego_terminado:
    return 

  boton_especifico = botones_grid[(r, c)]

  if es_bomba[(r, c)]:
    juego_terminado = True
    boton_especifico.config(text="游눢", bg="red", state="disabled", relief="sunken")
    
    for (fila, col), boton in botones_grid.items():
      if es_bomba[(fila, col)]:
        boton.config(text="游눢", bg="red")
  else:
    boton_especifico.config(state="disabled", relief="sunken", bg="lightgray")


def cronometro():
  global inicio
  if juego_terminado:
    # (Tu l칩gica de juego terminado aqu칤)
        # Cuidado: crear un Label aqu칤 cada segundo es un error.
        # Deber칤as tener UNA etiqueta de 'game over' y solo cambiar su texto.
    return 
    
  tiempo_transcurrido = time.time() - inicio
  string_tiempo = time.strftime('%H:%M:%S', time.gmtime(tiempo_transcurrido))
  etiqueta_cronometro.config(text=f"Cron칩metro {string_tiempo}")
  ventana.after(1000, cronometro) 


# --- PASO 1: HEMOS MOVIDO CASI TODO AQU칈 ---
def iniciar_juego():
    """Crea y muestra la pantalla principal del juego."""
    
    global inicio, juego_terminado, botones_grid, es_bomba, etiqueta_cronometro
    
    # Reiniciar variables del juego (por si se juega de nuevo)
    botones_grid = {}
    es_bomba = {}
    juego_terminado = False
    
    # 1. Ocultar el men칰 de opciones (que crearemos en el Paso 2)
    # marco_opciones.pack_forget() # Lo a침adiremos luego

    # 2. Crear los marcos del juego (como ten칤as antes)
    marco_etiquetas = tk.Frame(ventana)
    marco_etiquetas.pack(expand=True, anchor="n", pady=10)

    marco_juego = tk.Frame(ventana)
    marco_juego.configure(width=1, height=1)
    marco_juego.pack(pady=30)

    # 3. Crear la grilla de botones (como ten칤as antes)
    for i in range(TAMANO_GRID):
      for j in range(TAMANO_GRID):
        boton = tk.Button(marco_juego, text=" ", width=4, height=3)
        boton.grid(column=j, row=i)
        botones_grid[(i, j)] = boton
        es_bomba[(i, j)] = False
        boton.config(command=lambda r=i, c=j: casilla_presionada(r, c))

    # 4. Crear las bombas (como ten칤as antes)
    coordenadas_posibles = list(botones_grid.keys())
    coordenadas_bombas = random.sample(coordenadas_posibles, NUM_BOMBAS)
    for (r, c) in coordenadas_bombas:
      es_bomba[(r, c)] = True
    
    # 5. Crear y empezar el cron칩metro (como ten칤as antes)
    etiqueta_cronometro = tk.Label(marco_etiquetas, text="Cron칩metro", font=("Arial", 14))
    etiqueta_cronometro.pack()
    
    inicio = time.time()
    cronometro()

# --- C칍DIGO PRINCIPAL (Lo que se ejecuta al inicio) ---

ventana = tk.Tk()
ventana.title("Buscaminas")
ventana.geometry("430x720+1100+0")
ventana.minsize(430, 720)
ventana.maxsize(500, 720)
# (Aqu칤 ir칤a tu iconbitmap)


# --- PASO 2: CREAR LA PANTALLA DE OPCIONES ---
# (Esto se ejecuta AHORA, en lugar del juego)

marco_opciones = tk.Frame(ventana, pady=50) # Un marco para el men칰
marco_opciones.pack(expand=True)

etiqueta_titulo = tk.Label(marco_opciones, text="BUSCAMINAS", font=("Arial", 24, "bold"))
etiqueta_titulo.pack(pady=20)

# (Aqu칤 podr칤amos a침adir widgets para cambiar TAMANO_GRID, etc.)
etiqueta_tamano = tk.Label(marco_opciones, text="Tama침o (ej: 10):", font=("Arial", 12))
etiqueta_tamano.pack()
entrada_tamano = tk.Entry(marco_opciones, font=("Arial", 12), width=5)
entrada_tamano.pack(pady=5)
entrada_tamano.insert(0, "10") # Valor por defecto

etiqueta_bombas = tk.Label(marco_opciones, text="N췈 Bombas (ej: 10):", font=("Arial", 12))
etiqueta_bombas.pack()
entrada_bombas = tk.Entry(marco_opciones, font=("Arial", 12), width=5)
entrada_bombas.pack(pady=5)
entrada_bombas.insert(0, "10") # Valor por defecto


# --- Bot칩n para empezar ---
# Este bot칩n llamar치 a una *nueva* funci칩n que lee los valores y luego inicia el juego
def leer_opciones_y_jugar():
    global TAMANO_GRID, NUM_BOMBAS
    
    # 1. Leer valores de las cajas de texto
    TAMANO_GRID = int(entrada_tamano.get())
    NUM_BOMBAS = int(entrada_bombas.get())
    
    # 2. Ocultar este men칰
    marco_opciones.pack_forget()
    
    # 3. 춰Llamar a la funci칩n que crea el juego!
    iniciar_juego()


boton_jugar = tk.Button(marco_opciones, text="JUGAR", font=("Arial", 14, "bold"), command=leer_opciones_y_jugar)
boton_jugar.pack(pady=40, ipady=10, ipadx=20) # ipady/ipadx hacen el bot칩n m치s grande


ventana.mainloop()
```
